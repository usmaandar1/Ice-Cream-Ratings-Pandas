```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\usmaa\Documents\Just IT data tech bootcamp Dec 2023 - Mar 2024\Project Datasets - see onedrive too\Python Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#We create visualisations using plot 
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df.plot(kind = 'line')
```

```python
# The above shows a line graph but we want to create subplots to show the individual graphs for each ice cream rating
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
# Re plotting the same line graph but with a title 
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
df.plot(kind = 'line', subplots = True, title = 'Ice Cream Ratings')
```

```python
# plot a line graph showing the overall rating against the daily rating
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel= 'Daily Rating', ylabel = 'overall rating')
```

```python
# plotting a bar graph
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings')
```

```python
# plotting a stacked bar graph using the ratings
```

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings')
```

```python
#Plotting only the flavour rating on a bar graph
```

```python
df['Flavor Rating'].plot(kind = 'bar', title = 'Ice Cream Ratings')
```

```python
#Plotting the same stacked bar graph as earlier but turning it into a horizontal bar graph
```

```python
df.plot.barh(stacked = True)
```

```python
# Plotting a scatter graph containing overall and texture ratings
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'red')
```

```python
#Using bins to create a histogram of ice cream ratings
```

```python
df.plot.hist(bins = 10, title = 'Ice Cream Ratings')
```

```python
#plotting using a box plot
```

```python
df.plot.box()
```
